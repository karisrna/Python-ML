```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, neighbors, neural_network, model_selection, metrics
```

```python
data=datasets.load_iris()
X,y=data.data, data.target
```

```python
X=np.column_stack([X,X**2]) #(加入二次项)
```

```python
X.shape #变成2×2的矩阵，可提高精度
```

```python
model=linear_model.LogisticRegression()
model.fit(X,y)
y_predict=model.predict(X)
metrics.confusion_matrix(y,y_predict)
```

```python
model_selection.cross_val_score(model,X,y,cv=5)
```
