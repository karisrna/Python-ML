```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, neighbors, datasets, neural_network, model_selection, metrics, linear_model
```

```python
X,y=data=datasets.make_blobs(100,2,centers=2,cluster_std=1,random_state=3)
```

```python
X,y=data=datasets.make_blobs(100,2,centers=2,cluster_std=1,random_state=3)

model=svm.SVC(kernel='linear',C=100)
model.fit(X,y)
y_predict=model.predict(X)

fig,ax=plt.subplots()
ax.scatter(X[:,0],X[:,1],c=y,cmap='spring')
xs=np.linspace(ax.get_xlim()[0],ax.get_xlim()[1],100)
ys=np.linspace(ax.get_ylim()[0],ax.get_ylim()[1],100)
Z=np.zeros((100,100))
for ix in range(100):
    for iy in range(100):
        Z[ix,iy]=model.decision_function([[xs[ix],ys[iy]]])
ax.contour(xs,ys,Z,levels=[-1,0,1],linecolors='r',linestyles=['--','-','--'])
```
